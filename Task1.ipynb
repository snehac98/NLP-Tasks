```python
from nltk.corpus import wordnet as wn
word=input('Enter the word to search : ')
a=wn.synsets(word)
for i in a:
    print(i.lemma_names()[0],':',i.definition())
    print('Synonyms :',i.lemma_names())
```

```python
texts=["""Hello my name is pratyush, Currently im studying in 3rd year B.tech CSE in VIT vellore."""] #between 3 quotes

for text in texts:
    sentences=nltk.sent_tokenize(text)
    for sentence in sentences:
        word=nltk.word_tokenize(sentence)
        tagged_words= nltk.pos_tag(word)
        print(tagged_words)
```
