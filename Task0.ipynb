```python
from nltk.corpus import inaugural
inaugural.fileids()
```

```python
import collections
import math
from collections import Counter
```

```python
vec1=Counter(inaugural.words(fileids='2009-Obama.txt'))
vec2=Counter(inaugural.words(fileids='1793-Washington.txt'))
vec1
```

```python
intersection = set(vec1.keys()) & set(vec2.keys())
numerator = sum([vec1[x] * vec2[x] for x in intersection])
sum1 = sum([vec1[x]**2 for x in vec1.keys()])
sum2 = sum([vec2[x]**2 for x in vec2.keys()])
denominator = math.sqrt(sum1) * math.sqrt(sum2)
if not denominator:
    cosine=0.0
else:
    cosine=float(numerator) / denominator
print(cosine)
```

```python
brown.words(categories="mystery")[0:100]
```

```python
import nltk
cfd=nltk.ConditionalFreqDist(
    (genre,word)
    for genre in brown.categories()
    for word in brown.words(categories=genre))
genre=['religion','hobbies','mystery']
modals=['can','could','may']
cfd.tabulate(conditions=genre,samples=modals)
```
